# Predictive Churn Model

### Import the packages and modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick
import missingno as mn

sns.set(style='white')
```

### Read the dataset

```python
dataset = pd.read_excel('Churn.xlsx')
dataset.head()
```

```python
dataset.columns.values
```

### Check missing values

```python
dataset.isnull().sum()
```

Shows us TotalCharges has missing values

```python
mn.matrix(dataset)  # just for basic visualisation
```

```python
# Remove the missing values
dataset.dropna(inplace=True)

#removing the customer ID as it will not help in prediction
df = dataset
df = dataset.iloc[:, 1:]
df.head()
dataset.head()
```

```python
# Convertin the predictor variable in a binary numeric variable
# Didnt use Label Encoder, just used replace here
df['Churn'].replace(to_replace='Yes', value=1, inplace=True)
df['Churn'].replace(to_replace='No',  value=0, inplace=True)
```

```python
#Let's convert all the categorical variables into dummy variables
# As dummy variables will help the models we will use a great help
df_dummies = pd.get_dummies(df)
df_dummies.head()
```

```python
# Checking for collinearity
```

```python
# The chart of correlation of the target with other variables
df_dummies.corr()['Churn'].sort_values(ascending = False)
```

```python
# Get Correlation of "Churn" with other variables:
plt.figure(figsize=(15,8))
df_dummies.corr()['Churn'].sort_values(ascending = False).plot(kind='bar')
```

>We can see Month to Month Contract, Online Backup, Tech support, and device protection's absence, using fibre optics for internet connection made people to do more churning.

>Two year and One year Contract, the Total Charges paced, Customers with dependendents and partners are not churning

> A paperless billings also made an impact

## Part of the predictive models
I am going to use Logistic Classifier, Random Classifier, XGBoost and SVM as well

### A few steps of cleaning

```python
# We are going to use the dummy dataframe we have created
y = df_dummies['Churn'].values
X = df_dummies.drop(columns = ['Churn'])
```

```python
# Scaling using Min-Max Scaller
from sklearn.preprocessing import MinMaxScaler
features = X.columns.values
scaler = MinMaxScaler(feature_range = (0,1))
scaler.fit(X)
X = pd.DataFrame(scaler.transform(X))
X.columns = features
```

```python
# Creating training and test dataset
# As per requirement, 4000 in train and 3043 in test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.43, random_state=101)
```

## Logistic Regression

```python
# Running logistic regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
result = model.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

y_pred = model.predict(X_test)

print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# To get the weights of all the variables
weights = pd.Series(model.coef_[0],
                 index=X.columns.values)
print (weights.sort_values(ascending = False)[:10].plot(kind='bar', color='#F54748'))
```

```python
print(weights.sort_values(ascending = False)[-10:].plot(kind='bar', color='#00EAD3'))
```

### Hypothesis
> As we saw in our EDA, having a 2 month contract reduces chances of churn. 2 month contract along with tenure have the most negative relation with Churn as predicted by logistic regressions

> Having DSL internet service also reduces the proability of Churn

> Lastly, total charges, monthly contracts, fibre optic internet services and seniority can lead to higher churn rates. This is interesting because although fibre optic services are faster, customers are likely to churn because of it. I think we need to explore more to better understad why this is happening.

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(n_estimators=1000 , oob_score = True, n_jobs = -1,
                                  random_state =50, max_features = "auto",
                                  max_leaf_nodes = 30)
model_rf.fit(X_train, y_train)

# Make predictions
y_pred = model_rf.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
importances = model_rf.feature_importances_
weights = pd.Series(importances,
                 index=X.columns.values)
weights.sort_values()[-10:].plot(kind = 'bar')
```

#### Hypothesis

> From random forest algorithm, monthly contract, tenure and total charges are the most important predictor variables to predict churn.
> The results from random forest are very similar to that of the logistic regression.

## XGBoost

```python
from xgboost import XGBClassifier
model_xg = XGBClassifier()
model_xg.fit(X_train, y_train)

# Make predictions
y_pred = model_xg.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

## Support Vector Machine

```python
from sklearn.svm import SVC

model_svm = SVC(kernel='linear') 
model_svm.fit(X_train,y_train)

# Make predictions
y_pred = model_svm.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
